### Comprensión de listas, Generadores e Itertools

Introducción a técnicas avanzadas para la creación, iteración y modificación de listas y diccionarios.

La comprensión de listas es una técnica que nos permite crear listas o diccionarios de manera comprensible y sencilla. A continuación se presentan dos formas de crear una lista: usando _comprension de listas_ o usando append.

```python
# Append
lista_a = []
for i in range(10):
    lista_a.append(i**2)
    
# Comprension de listas
lista_c = [i**2 for i in range(10)]
```

#### Sintaxis
En general se sigue la siguiente sintanxis para las listas:

```python
# Por valores en un iterable
l1 = [i for i in range(100)]

# Condicionales
l2 = [i/2 for i in range(100) if i%2==0]
l3 = [1 if i%2==0 else 0 for i in range(100) ]
```

Y para los diccionarios

```python
# Por valores en un iterable
d1 = { index:value for index, value in enumerate("abcdefghi") if value not in "aeiou"}
d2 = {l1_i : l3_i for l1_i, l3_i in zip(l1, l3)}
```

### Funciones de orden superior e itertool

Además tenemos a disposición elementos del paradigma funcional que nos ayudan a filtrar y manipular iterables

### Lambda

La función anónima lambda nos permite definir funciones sencillas. Esto nos ahorra tener que definir en un bloque de código filtros simples o funciones que solo se utilizan una vez.

Por ejemplo necesitamos las funciones $f(x) = x + 2$ o un comprar dos números:

```python
# Guardar la función en una variable
f = lambda x: x+2
print(f(4))

# Definir y llamar la función 
es_mayor = (lambda a,b: a >= b)(3, 5)
```

### Map

_map_ es un generador que aplica una función a cada elemento de iterable:

```python
potencias = list(map(lambda n**2, range(100))
```

### Filter

_filter_ filtra el iterable para aquellos valores que cumplan la condición:

```python
consonantes = list(filter(lambda c: c not in "aeiou", "sebastian"))
```

### Reduce
_reduce_ nos ayuda a realizar computos sobre los iterables:

```python
from functools import reduce
producto = reduce((lambda x, y: x * y), [1, 2, 3, 4])
```
